```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

This is a Simple Budget project, where using the Pandas library, I will create a budget and analyze the data predicting how mush money I will have in the future.

```python
# Defining Indexes Columns and expenses.
my_indexes = ["Car", "Housing", "DayCare", "ExtraActivities", "Food", "Fuel", "Extra"]
my_columns = ["Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
expenses = {
    "Car": 209,
    "Housing": 400,
    "DayCare": 57,
    "ExtraActivities": 10,
    "Food": 300,
    "Fuel": 70,
    "Extra": 100
}

data = {column: [expenses[index] for index in my_indexes] for column in my_columns}

# Creating the DataFrame
budget = pd.DataFrame(data , index=my_indexes, columns=my_columns)
```

```python
budget
```

Adding a Total Expense Row to the DataFrame, I will be able to see how much money 
I will spend in the future every month.

```python
total=[]
for col in budget.columns:
    total.append(budget[col].sum())
total
```

```python
total_data= dict(zip(my_columns, total))
total_data
```

```python
budget.loc["Total"] = total_data
```

```python
budget
```

Now let's create the income DataFrame.

```python
income_Values={
    "Salary": 856,
    "Extra": 140,
    "Tip": 100,
    "Others": 50
}

income_indexes=["Salary", "Extra", "Tip", "Others"]
monthly_columns=["Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

income_data={column: [income_Values[index] for index in income_indexes] for column in monthly_columns}

#Creating the DataFrame
income=pd.DataFrame(income_data, index=income_indexes, columns=monthly_columns)
income
```

I forgot to include extra income that just happend two months a year. I will add this to the DataFrame.

```python
extra_income={
    "Jun": 0,
    "Jul": 996,
    "Aug": 0,
    "Sep": 0,
    "Oct": 0,
    "Nov": 0,
    "Dec": 996
}

income.loc["HollidaysAndChristmas"]=extra_income
income
```

Adding a total row to the income DataFrame, I will be able to see how much money I will have in the future every month.

```python
totalrow= dict(zip(monthly_columns, income.sum()))
totalrow
```

Adding the total row to the income DataFrame:

```python
income.loc["Total"]=totalrow
income
```

What we have so far:
Income: The total income I get monthly.
Budget: The expenses I have monthly.

```python
income
```

```python
budget
```

I'm going to rename the total columns, so we can identify income and expenses.

```python
income.rename(index={"Total": "Total Income"}, inplace=True)
budget.rename(index={"Total": "Total Expenses"}, inplace=True)
```

Now I going to combine the two dataframes into one, and I will use the method concat() to do this.

```python
budget
```

```python
income
```

```python
my_budget=pd.concat([income, budget], axis=0)
my_budget
```

```python
import matplotlib.pyplot as plt
```

Now let´s compare the income and expenses, and see how much money I will have in the future:

```python
visual_income=my_budget.loc["Total Income"]
visual_expenses=my_budget.loc["Total Expenses"]
plt.figure(figsize=(10, 6))  # Adjust the size of the plot
plt.plot(visual_income.index, visual_income, marker='o', label='Income')  # Income plot
plt.plot(visual_expenses.index, visual_expenses, marker='s', label='Expenses')  # Expenses plot

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Values')
plt.title('Income and Expenses per month.')
plt.xticks(rotation=45)  # Rotate the x labels for better visualization
plt.legend()

# Show the plot
plt.show()
```

Now let's calculate the balance for each month and the total balance.

```python
balance = income.loc["Total Income"] - budget.loc["Total Expenses"]
balance
```

And the final total balance is:

```python
balance.sum()
```
